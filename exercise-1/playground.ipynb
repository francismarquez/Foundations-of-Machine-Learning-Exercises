```python
import numpy as np  
import matplotlib.pyplot as plt
```

```python
def positional_encoding(position, d_model):
  steps = np.linspace(0,2*np.pi,(d_model//2))
  angle_rads = []

  for x in range(len(steps)): 
    angle_rads.append(steps[x])
    angle_rads.append(steps[x])

  angle_rads = np.array(angle_rads)
  angle_rads = np.repeat(angle_rads[:,np.newaxis], d_model,1).T

  tokens = np.arange(position)
  print(tokens)
  tokens = np.sqrt(tokens)
  print(tokens)
  dim = np.arange(d_model)
  dim = np.power(np.sqrt(2),dim/d_model)

  pos = np.repeat(tokens[:,np.newaxis], position,1)
  dim = np.repeat(dim[:,np.newaxis], d_model,1).T

  pos_dim = np.true_divide(pos,dim)
  angle_rads = np.multiply(angle_rads,pos_dim)

  angle_rads[:, 0::2] = np.sin(angle_rads[:, 0::2])
  angle_rads[:, 1::2] = np.cos(angle_rads[:, 1::2]) 

  position_encoding = angle_rads[np.newaxis, ...]
  return position_encoding
```

```python
position = 16
dimensions = 16

pos_encoding = positional_encoding(position, dimensions)

plt.figure(figsize=(6,6))
plt.pcolormesh(pos_encoding[0], cmap='RdBu')
plt.xlabel('Depth')
plt.xlim((0, dimensions))
plt.ylim((position,0))
plt.ylabel('Position')
plt.colorbar()
plt.show()
```
